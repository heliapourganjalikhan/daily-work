```python
class Node :
    def __init__ (self ,data ,nextval = None):
        self.data = data
        self.nextval = nextval

class Slinkedlist :
    def __init__(self):
        self.head = None
        
    #traversing the linked list
    def Printlist (self) :
        printval = self.head
        while printval is not None:
            print(printval.data, end ="")
            printval= printval.nextval
    
    #insertion of a new node in the beginig of the list
    def AtBegining (self ,newdata):
        NewNode = Node(newdata)
        NewNode.nextval = self.head
        print(self.head)
        self.head = NewNode
        print(self.head)
        
    #insertion at the end of the list
    def AtEnd (self ,newnode):
        NewNode = Node(newnode)
        if self.head is None:
            self.head = NewNode
            return
        tail = self.head
        while(tail.nextval):
            tail = tail.nextval
        tail.nextval = NewNode
        
    #insertion in the middle of list
    def InBetween (self ,middle_node ,newdata):
        Newnode = Node(newdata)
        if middle_node is None:
            print("This middle node is absent")
            return
        Newnode.nextval = middle_node.nextval
        middle_node.nextval = Newnode
        
    #remove node
    #there are three main senarios for removing an item from a linked list :
        # 1) the removekey is in the first node where the head of the list should change
        # 2) the removekey is not in the first node so it might be in the middle of the list
        #or even be absent;in this case we need to track the previous node of the node which 
        #is to be checked
        # 3) the linked list is empty
    def RemoveNode (self ,removekey):
        Head = self.head
        
        if Head is not None:
            if Head.data == removekey:
                self.head = Head.nextval
                Head = None
                return
        
        while Head is not None:
            if Head.data == removekey:
                break
            prev = Head
            Head = Head.nextval
        if Head is None:
            return
        prev.nextval = Head.nextval
        Head = None
        
        
list1 = Slinkedlist()
list1.head = Node(1)
digit2 = Node(2)
digit3 = Node(3)
#create links between nodes
list1.head.nextval = digit2
digit2.nextval = digit3
print(list1.head.data)
list1.AtBegining(4)
list1.AtEnd(4)
print(list1.head)
list1.InBetween(digit2.nextval,3)
list1.RemoveNode(2)
list1.Printlist()
```
